FundMapping
Relate a few indices to a fund.
In Mathematical terms come up with a weights that relate the returns of each indices $ A_1,A_2, ..., A_n $ with and equity Fund $F$
$$ L_i(k) = log\left(\frac{A_i(k+1)}{A_i(k)}\right) k = 1, ..., m-1, i = 1, n $$
$$LRF(k) = log\left(\frac{F(k+1)}{F(k)}\right) k = 1, .., m-1 $$

$$ [L_1, L_2, ... , L_n]  \left[ \begin{array}{ccc}
\beta_1&\\
\vdots&\\
\beta_n&\end{array} \right]  
= [LRF]$$
